```python
import pandas as pd
```

```python
with open("../slamdunk/SRR_Acc_List.txt", "r") as f:
    sampleID = f.read().splitlines()
    print(sampleID)
```

```python
for ID in sampleID:
    path = f"../slamdunk/count/{ID}/{ID}_1.fastq_slamdunk_mapped_filtered_tcount.tsv"
    print(path)
```

## ReadCount のみを集めた allreads.csv ファイルを作成

```python
column = "ReadCount"

path = f"../slamdunk/count/{sampleID[0]}/{sampleID[0]}_1.fastq_slamdunk_mapped_filtered_tcount.tsv"
df =  pd.read_csv(path, sep="\t", index_col=["Chromosome", "Start", "End", "Name", "Length", "Strand"], skiprows=2)
df = df.loc[:, [column]]
df = df.rename(columns={column: sampleID[0]})

for ID in sampleID[1:]:
    path = f"../slamdunk/count/{ID}/{ID}_1.fastq_slamdunk_mapped_filtered_tcount.tsv"    
    tmp = pd.read_csv(path, sep="\t", index_col=["Chromosome", "Start", "End", "Name", "Length", "Strand"], skiprows=2)
    df = pd.merge(df, tmp[column], left_index=True, right_index=True, how="outer")
    df = df.rename(columns={column: ID})
    
df
```

```python
df.to_csv("../slamdunk/csv/allreads.csv")
```

## TcReadCount のみを集めた tcreads.csv ファイルを作成

```python
column = "TcReadCount"

path = f"../slamdunk/count/{sampleID[0]}/{sampleID[0]}_1.fastq_slamdunk_mapped_filtered_tcount.tsv"
df =  pd.read_csv(path, sep="\t", index_col=["Chromosome", "Start", "End", "Name", "Length", "Strand"], skiprows=2)
df = df.loc[:, [column]]
df = df.rename(columns={column: sampleID[0]})

for ID in sampleID[1:]:
    path = f"../slamdunk/count/{ID}/{ID}_1.fastq_slamdunk_mapped_filtered_tcount.tsv"    
    tmp = pd.read_csv(path, sep="\t", index_col=["Chromosome", "Start", "End", "Name", "Length", "Strand"], skiprows=2)
    df = pd.merge(df, tmp[column], left_index=True, right_index=True, how="outer")
    df = df.rename(columns={column: ID})
    
df
```

```python
df.to_csv("../slamdunk/csv/tcreads.csv")
```

### Total reads から T>C reads を差し引いた non-T>C reads のテーブルを作る

```python
allreads = pd.read_csv("../slamdunk/csv/allreads.csv", index_col=["Chromosome", "Start", "End", "Name", "Length", "Strand"])
tcreads = pd.read_csv("../slamdunk/csv/tcreads.csv", index_col=["Chromosome", "Start", "End", "Name", "Length", "Strand"])

allreads.head()
```

```python
ntcreads = allreads - tcreads
```

```python
ntcreads.to_csv("../slamdunk/csv/ntcreads.csv")
```
