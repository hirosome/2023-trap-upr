```python
import pandas as pd
```

```python
allreads = pd.read_csv("../slamdunk/csv/allreads.csv", index_col="gene_name")
tcreads = pd.read_csv("../slamdunk/csv/tcreads.csv", index_col="gene_name")

allreads = allreads.filter(regex=("Total|P0"), axis=1)
tcreads = tcreads.filter(regex=("Total|P0"), axis=1)

allreads.head()
```

```python
column_names = [ "RIBO-DMSO-1", "RIBO-DMSO-2", "RIBO-DMSO-3", "RNA-DMSO-1", "RNA-DMSO-2", "RNA-DMSO-3",
                                "RIBO-TPG-1", "RIBO-TPG-2", "RIBO-TPG-3", "RNA-TPG-1", "RNA-TPG-2", "RNA-TPG-3"]
```

```python
allreads.columns = column_names
tcreads.columns = column_names
```

```python
allreads.head()
```

```python
allreads.filter(regex="RIBO", axis=1).to_csv("../rscript/deltaTE/all/input_txt/P-TRAP_DMSOvsTPG_All.txt", sep="\t")
allreads.filter(regex="RNA", axis=1).to_csv("../rscript/deltaTE/all/input_txt/Total_DMSOvsTPG_All.txt", sep="\t")
```

```python
tcreads.filter(regex="RIBO", axis=1).to_csv("../rscript/deltaTE/tc/input_txt/P-TRAP_DMSOvsTPG_TC.txt", sep="\t")
tcreads.filter(regex="RNA", axis=1).to_csv("../rscript/deltaTE/tc/input_txt/Total_DMSOvsTPG_TC.txt", sep="\t")
```
