```python
import pandas as pd
```

```python
samples = ["DMSO-Total-1", "DMSO-Total-2", "DMSO-Total-3", "TPG-Total-1", "TPG-Total-2", "TPG-Total-3",
                    "DMSO-P0-1", "DMSO-P0-2", "DMSO-P0-3", "TPG-P0-1", "TPG-P0-2", "TPG-P0-3",
                    "DMSO-L10a-1", "DMSO-L10a-2", "DMSO-L10a-3", "TPG-L10a-1", "TPG-L10a-2", "TPG-L10a-3"]
```

```python
allreads = pd.read_csv("../slamdunk/csv/allreads.csv", index_col=["Chromosome", "Name"]).drop(["Start", "End", "Length", "Strand"], axis=1)
tcreads = pd.read_csv("../slamdunk/csv/tcreads.csv", index_col=["Chromosome", "Name"]).drop(["Start", "End", "Length", "Strand"], axis=1)

allreads.columns = samples
tcreads.columns = samples

allreads.head()
```

```python
allreads = allreads.reset_index()
allreads["Gene"] = allreads["Name"] + "|" + allreads["Chromosome"]
allreads = allreads.set_index("Gene")
allreads = allreads.drop(columns=["Name", "Chromosome"])

tcreads = tcreads.reset_index()
tcreads["Gene"] = tcreads["Name"] + "|" + tcreads["Chromosome"]
tcreads = tcreads.set_index("Gene")
tcreads = tcreads.drop(columns=["Name", "Chromosome"])
```

```python
allreads= allreads.filter(regex="Total|P0").sort_index(axis=1)
tcreads= tcreads.filter(regex="Total|P0").sort_index(axis=1)
```

```python
allreads.head()
```

```python
column_names = [ "RIBO-DMSO-1", "RIBO-DMSO-2", "RIBO-DMSO-3", "RNA-DMSO-1", "RNA-DMSO-2", "RNA-DMSO-3",
                                "RIBO-TPG-1", "RIBO-TPG-2", "RIBO-TPG-3", "RNA-TPG-1", "RNA-TPG-2", "RNA-TPG-3"]
```

```python
allreads.columns = column_names
tcreads.columns = column_names
```

```python
allreads.head()
```

```python
allreads.filter(regex="RIBO", axis=1).to_csv("../rscript/deltaTE_v2/all/input_txt/P-TRAP_DMSOvsTPG_All.txt", sep="\t")
allreads.filter(regex="RNA", axis=1).to_csv("../rscript/deltaTE_v2/all/input_txt/Total_DMSOvsTPG_All.txt", sep="\t")
```

```python
tcreads.filter(regex="RIBO", axis=1).to_csv("../rscript/deltaTE_v2/tc/input_txt/P-TRAP_DMSOvsTPG_TC.txt", sep="\t")
tcreads.filter(regex="RNA", axis=1).to_csv("../rscript/deltaTE_v2/tc/input_txt/Total_DMSOvsTPG_TC.txt", sep="\t")
```
