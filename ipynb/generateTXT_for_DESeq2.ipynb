```python
import pandas as pd
```

```python
samples = ["DMSO-Total-1", "DMSO-Total-2", "DMSO-Total-3", "TPG-Total-1", "TPG-Total-2", "TPG-Total-3",
                    "DMSO-P0-1", "DMSO-P0-2", "DMSO-P0-3", "TPG-P0-1", "TPG-P0-2", "TPG-P0-3",
                    "DMSO-L10a-1", "DMSO-L10a-2", "DMSO-L10a-3", "TPG-L10a-1", "TPG-L10a-2", "TPG-L10a-3"]
```

```python
allreads = pd.read_csv("../slamdunk/csv/allreads.csv", index_col=["Chromosome", "Name"]).drop(["Start", "End", "Length", "Strand"], axis=1)
tcreads = pd.read_csv("../slamdunk/csv/tcreads.csv", index_col=["Chromosome", "Name"]).drop(["Start", "End", "Length", "Strand"], axis=1)

allreads.columns = samples
tcreads.columns = samples

allreads.head()
```

row.names が重複していると DESeq2 でエラーとなるので Chromosome の情報も入れる。

```python
allreads = allreads.reset_index()
allreads["Gene"] = allreads["Name"] + "|" + allreads["Chromosome"]
allreads = allreads.set_index("Gene")
allreads = allreads.drop(columns=["Name", "Chromosome"])

tcreads = tcreads.reset_index()
tcreads["Gene"] = tcreads["Name"] + "|" + tcreads["Chromosome"]
tcreads = tcreads.set_index("Gene")
tcreads = tcreads.drop(columns=["Name", "Chromosome"])
```

```python
allreads.filter(regex="Total", axis=1).to_csv("../rscript/deseq2_v2/input_txt/Total_DMSOvsTPG_All.txt", sep="\t")
allreads.filter(regex="P0", axis=1).to_csv("../rscript/deseq2_v2/input_txt/P-TRAP_DMSOvsTPG_All.txt", sep="\t")
allreads.filter(regex="L10a", axis=1).to_csv("../rscript/deseq2_v2/input_txt/L10a-TRAP_DMSOvsTPG_All.txt", sep="\t")
```

```python
tcreads.filter(regex="Total", axis=1).to_csv("../rscript/deseq2_v2/input_txt/Total_DMSOvsTPG_TC.txt", sep="\t")
tcreads.filter(regex="P0", axis=1).to_csv("../rscript/deseq2_v2/input_txt/P-TRAP_DMSOvsTPG_TC.txt", sep="\t")
tcreads.filter(regex="L10a", axis=1).to_csv("../rscript/deseq2_v2/input_txt/L10a-TRAP_DMSOvsTPG_TC.txt", sep="\t")
```

```python
allreads.filter(regex="DMSO-P0|DMSO-L10a", axis=1).to_csv("../rscript/deseq2_v2/input_txt/P0vsL10a_DMSO_All.txt", sep="\t")
tcreads.filter(regex="DMSO-P0|DMSO-L10a", axis=1).to_csv("../rscript/deseq2_v2/input_txt/P0vsL10a_DMSO_TC.txt", sep="\t")
```

Non-T>C reads についても解析する

```python
ntcreads = pd.read_csv("../slamdunk/csv/ntcreads.csv", index_col=["Chromosome", "Name"]).drop(["Start", "End", "Length", "Strand"], axis=1)
ntcreads.columns = samples
```

```python
ntcreads = ntcreads.reset_index()
ntcreads["Gene"] = ntcreads["Name"] + "|" + ntcreads["Chromosome"]
ntcreads = ntcreads.set_index("Gene")
ntcreads = ntcreads.drop(columns=["Name", "Chromosome"])
```
