```python
import pandas as pd
```

```python
allreads = pd.read_csv("../slamdunk/csv/allreads.csv", index_col="gene_name")
tcreads = pd.read_csv("../slamdunk/csv/tcreads.csv", index_col="gene_name")

allreads.head()
```

```python
allreads.filter(regex="Total", axis=1).to_csv("../rscript/deseq2/input_txt/Total_DMSOvsTPG_All.txt", sep="\t")
allreads.filter(regex="P0", axis=1).to_csv("../rscript/deseq2/input_txt/P-TRAP_DMSOvsTPG_All.txt", sep="\t")
allreads.filter(regex="L10a", axis=1).to_csv("../rscript/deseq2/input_txt/L10a-TRAP_DMSOvsTPG_All.txt", sep="\t")
```

```python
tcreads.filter(regex="Total", axis=1).to_csv("../rscript/deseq2/input_txt/Total_DMSOvsTPG_TC.txt", sep="\t")
tcreads.filter(regex="P0", axis=1).to_csv("../rscript/deseq2/input_txt/P-TRAP_DMSOvsTPG_TC.txt", sep="\t")
tcreads.filter(regex="L10a", axis=1).to_csv("../rscript/deseq2/input_txt/L10a-TRAP_DMSOvsTPG_TC.txt", sep="\t")
```
